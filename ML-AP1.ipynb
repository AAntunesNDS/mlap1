```python
featureWithNullValues = data.columns[data.isna().any()].tolist()
len(featureWithNullValues)
```

```python
for i in featureWithNullValues:
    data[i] = data[i].replace(np.nan, data[i].mean())
```

```python
data['label'] = pd.get_dummies(data['label'], columns=data['label'], prefix=data['label'], drop_first=True)
```
